THEORY QUESTIONS :

1. What does R-squared represent in a regression model

   - R-squared (R²) is a statistical measure that represents the proportion of
     variance in the dependent variable (Y) that is explained by the independent variables (X) in a regression model.
   - Interpretation:
      - R² = 0 → Model explains none of the variance.
      - R² = 1 → Model explains all the variance.
      - Higher R² means a better fit, but not always a better model (watch for
        overfitting).

2. What are the assumptions of linear regression ?

   - Assumptions of Linear Regression
      1. Linearity
      2. Independence
      3. Homoscedasticity
      4. Normality of Residuals
      5. No Multicollinearity (for multiple regression)

3. What is the difference between R-squared and Adjusted R-squared ?

   - R-squared (R²):
      - Proportion of variance explained by the model
      - Always increases or stays the same when adding predictors
      -	May give false sense of a good fit with many variables
      - Good for simple models

   - Adjusted R-squared:
      - R² adjusted for the number of predictors in the model
      - Increases only if new predictors improve the model
      - Penalizes unnecessary predictors, reducing overfitting risk
      - Better for comparing models with different numbers of predictors


4. Why do we use Mean Squared Error (MSE) ?

   - Mean Squared Error (MSE) is used to measure the average squared difference
     between actual and predicted values in regression models.
   - Quantifies prediction error
   - Penalizes large errors
   - Smooth and differentiable
   - Widely used for model evaluation


5.  What does an Adjusted R-squared value of 0.85 indicate ?

   - 85% of the variation in the dependent variable is explained by the
     independent variables, after adjusting for the number of predictors in the model
   - Why it's important:
      - Unlike regular R², adjusted R² accounts for model complexity.
      - A value of 0.85 suggests a very good fit, assuming the model isn't overfitting and assumptions are met.


6. How do we check for normality of residuals in linear regression ?

   - How to Check for Normality of Residuals in Linear Regression
   1. Histogram of Residuals
      - Plot residuals; they should look roughly bell-shaped (normal
        distribution).
   2. Q-Q Plot (Quantile-Quantile Plot)
      - Plots residual quantiles against theoretical normal quantiles.
      - If points lie along the diagonal, residuals are approximately normal.
   3. Shapiro-Wilk Test / Kolmogorov-Smirnov Test
      - Statistical tests to check if residuals follow a normal distribution.



7. What is multicollinearity, and how does it impact regression ?

  - Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other.
  - Impact on Regression:
   1. Unstable coefficients
     - Regression coefficients can change wildly with small changes in data.
   2. Difficult interpretation
     - It becomes hard to determine the individual effect of correlated
       predictors.
   3. Inflated standard errors
     - Makes it harder to assess the significance of predictors.


8. What is Mean Absolute Error (MAE)

   - Mean Absolute Error (MAE) is a metric used to measure the average absolute
     difference between actual and predicted values in regression.
   - Key Features:
      - Interpretable: Gives the average error in the same units as the target variable.
      - Less sensitive to outliers than MSE (no squaring).
      - Used to evaluate how close predictions are to actual outcomes.


9. What are the benefits of using an ML pipeline ?

   - Benefits of Using an ML Pipeline
      - Automation of Workflow
      - Consistency
      - Cleaner Code
      - Reduces Errors
      - Facilitates Experimentation
      - Supports Cross-Validation


10. What is pickling in Python, and how is it useful in ML ?

   - Pickling is the process of serializing (converting) a Python object into a
     byte stream so it can be saved to a file or transmitted and later restored using unpickling.
   - Why Its Useful in Machine Learning:
      - Save Trained Models
      - Efficient Deployment
      - Preserves Entire Pipeline
   - Pickling saves ML models and objects to disk, making it easy to reuse or
     deploy them without retraining.


11. What does a high R-squared value mean ?

   - A high R-squared value means:
      - The regression model explains a large proportion of the variance in the
        dependent variable.
      - The model fits the data well, capturing most of the underlying pattern.
      - Predictions are generally more accurate within the range of the  
        training data.
      - A high R² doesnt always mean the model is good—it can be high due to
        overfitting or irrelevant variables.
      - Always check other metrics and assumptions to confirm model quality.


12. What happens if linear regression assumptions are violated ?

   - If linear regression assumptions are violated, the consequences can
     include:
     1. Biased or inefficient estimates
     2. Invalid inference
     3. Poor predictions
     4. Heteroscedasticity
     5. Multicollinearity
     6. Non-normal residuals


13. How can we address multicollinearity in regression ?

   - Here are common ways to address multicollinearity in regression:
     1. Remove correlated variables
     2. Combine variables
        - Create composite features (e.g., average or principal components).
     3. Use Regularization
     4. Increase sample size
     5. Check data collection
        - Ensure predictors are measured accurately and distinctly.


14. How is Adjusted R-squared calculated ?

   - Adjusted R-squared modifies the regular R-squared to account for the  
     number of predictors in the model, penalizing unnecessary variables.
   - What it does:
     - If adding a predictor improves the model fit enough, adjusted R²
       increases.
     - If the new predictor doesn't help, adjusted R² decreases.


15. Why is MSE sensitive to outliers ?

    - MSE is sensitive to outliers because it squares the errors before
      averaging.
    - Squaring makes large errors (from outliers) much bigger compared to small
      errors.
    - This means a single outlier can disproportionately increase the MSE,
      skewing the overall error measure.
    - MSE’s squaring of errors amplifies outliers’ impact, making it sensitive
      to extreme values.

16. What is the role of homoscedasticity in linear regression ?

   - Homoscedasticity means the variance of residuals (errors) is constant across all levels of the independent variables.
   -Why it matters:
      - Ensures reliable standard errors of coefficients.
      - Validates hypothesis tests (t-tests, F-tests) and confidence intervals.
      - Helps the model's predictions have consistent accuracy across the data
        range.

  - If violated (heteroscedasticity):
     - Standard errors may be biased.
     - Test results can be misleading.

17. What is Root Mean Squared Error (RMSE) ?

   - RMSE is a metric that measures the average magnitude of the prediction
     errors in a regression model, expressed in the same units as the target variable.
   - Why use RMSE?
      - Penalizes large errors more than MAE because errors are squared before
        averaging.
      - Easier to interpret than MSE since it's in original units.
      - Commonly used to evaluate regression model accuracy.


18. Why is pickling considered risky ?

   - Pickling is considered risky mainly because:
   - Security Vulnerability
      - Loading (unpickling) data from untrusted sources can execute malicious
       code, potentially harming your system.
   - Lack of Compatibility
   - Not Human-Readable


19. What alternatives exist to pickling for saving ML models ?

   - Alternatives to Pickling for Saving ML Models
     1. Joblib
     2. ONNX (Open Neural Network Exchange)
     3. HDF5 (using h5py or Keras models)
     4. JSON / YAML
     5. Framework-specific formats


20. What is heteroscedasticity, and why is it a problem ?

   - Heteroscedasticity means the variance of the residuals (errors) is not
     constant across different levels of the independent variables — errors spread unevenly.
   - Why is it a Problem?
      - Violates a key assumption of linear regression (homoscedasticity).
      - Leads to biased standard errors, making hypothesis tests (t-tests,    F-tests) unreliable.
      - Confidence intervals and p-values become inaccurate.
      - Can reduce the efficiency of coefficient estimates.


21. How can interaction terms enhance a regression model's predictive power ?

   - Interaction terms capture the effect of two (or more) variables combined,
     showing how the relationship between one predictor and the outcome changes depending on another predictor.
   - Why it helps:
      - Models non-additive relationships between variables.
      - Captures more complex patterns in data.
      - Improves predictive accuracy by including variable dependencies.

```python
#  Write a Python script to visualize the distribution of errors (residuals) for a multiple linear regression model using Seaborn's "diamonds" dataset.

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

diamonds = sns.load_dataset('diamonds')

X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

residuals = y_test - y_pred

plt.figure(figsize=(10,6))
sns.histplot(residuals, kde=True, bins=30, color='skyblue')
plt.title('Distribution of Residuals (Errors)')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.show()
```

```python
#  Write a Python script to calculate and print Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE) for a linear regression model.

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import numpy as np
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
#  Write a Python script to check if the assumptions of linear regression are met. Use a scatter plot to check linearity, residuals plot for homoscedasticity, and correlation matrix for multicollinearity

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

diamonds = sns.load_dataset('diamonds')

X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

residuals = y_test - y_pred

plt.figure(figsize=(6, 4))
sns.scatterplot(x=X['carat'], y=y)
plt.title('Scatter plot: Carat vs Price (Linearity check)')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(6, 4))
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Predicted Values (Homoscedasticity check)')
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.show()

plt.figure(figsize=(6, 5))
corr_matrix = X.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Predictors (Multicollinearity check)')
plt.show()
```

```python
#  Write a Python script that creates a machine learning pipeline with feature scaling and evaluates the performance of different regression models

import seaborn as sns
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error

diamonds = sns.load_dataset('diamonds')

X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Lasso Regression': Lasso(alpha=0.1),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42)
}

for name, model in models.items():

    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', model)
    ])

    pipeline.fit(X_train, y_train)

    y_pred = pipeline.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    print(f"{name} RMSE: {rmse:.2f}")
```

```python
# Write a Python script that fits a polynomial regression model (degree 2) to a dataset and plots the regression curve.


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(0)
X = 6 * np.random.rand(100, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(100, 1)

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

X_curve = np.linspace(X.min(), X.max(), 100).reshape(100, 1)
X_curve_poly = poly.transform(X_curve)
y_curve = model.predict(X_curve_poly)

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_curve, y_curve, color='red', linewidth=2, label='Polynomial regression curve (degree 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()
```

```python
#  Write a Python script that creates a machine learning pipeline with data standardization and a multiple linear regression model, and prints the R-squared score

import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

diamonds = sns.load_dataset('diamonds')

X = diamonds[['carat', 'depth', 'table']]
y = diamonds['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

pipeline.fit(X_train, y_train)

y_pred = pipeline.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R-squared score: {r2:.4f}")
```

```python
# Write a Python script that performs multiple linear regression on a synthetic dataset with 5 features. Print the R-squared score and model coefficients

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

np.random.seed(42)
X = np.random.rand(200, 5)
true_coefs = np.array([1.5, -2.0, 3.0, 0.5, 4.0])
y = X @ true_coefs + np.random.randn(200) * 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R-squared score: {r2:.4f}")
print("Model coefficients:")
for i, coef in enumerate(model.coef_):
    print(f" Feature {i+1}: {coef:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
```

```python
#  Write a Python script that demonstrates how to serialize and deserialize machine learning models using joblib instead of pickling.

from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import joblib

X, y = make_regression(n_samples=100, n_features=2, noise=0.1, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

joblib.dump(model, 'linear_regression_model.joblib')
print("Model saved to 'linear_regression_model.joblib'")

loaded_model = joblib.load('linear_regression_model.joblib')
print("Model loaded from file")

y_pred = loaded_model.predict(X_test)
print(f"Predictions on test data: {y_pred[:5]}")
```

```python
# Write a Python script to perform linear regression with categorical features using one-hot encoding. Use the Seaborn 'tips' dataset.

import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

tips = sns.load_dataset('tips')

X = tips[['total_bill', 'sex', 'smoker', 'day', 'time']]
y = tips['tip']

X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print(f"R-squared score: {r2:.4f}")
print("Model coefficients:")
for feature, coef in zip(X_encoded.columns, model.coef_):
    print(f" {feature}: {coef:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
```

```python
# Compare Ridge Regression with Linear Regression on a synthetic dataset and print the coefficients and Rsquared score.

import numpy as np
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

np.random.seed(42)
X = np.random.rand(100, 5)
true_coefs = np.array([3, -1.5, 2, 0, 4])
y = X @ true_coefs + np.random.randn(100) * 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)

ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Linear Regression:")
print(f" Coefficients: {lr.coef_}")
print(f" R-squared: {r2_lr:.4f}\n")

print("Ridge Regression (alpha=1.0):")
print(f" Coefficients: {ridge.coef_}")
print(f" R-squared: {r2_ridge:.4f}")
```

```python
# Write a Python script that uses cross-validation to evaluate a Linear Regression model on a synthetic dataset.

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score

X, y = make_regression(n_samples=150, n_features=4, noise=10, random_state=42)

model = LinearRegression()

cv_scores = cross_val_score(model, X, y, cv=5, scoring='r2')

print("Cross-validation R-squared scores:", cv_scores)
print(f"Average R-squared: {cv_scores.mean():.4f}")
```

```python
# Write a Python script that compares polynomial regression models of different degrees and prints the Rsquared score for each

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

np.random.seed(42)
X = 6 * np.random.rand(200, 1) - 3
y = 0.5 * X**3 - X**2 + X + 2 + np.random.randn(200, 1) * 3

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

degrees = [1, 2, 3, 4, 5]

for degree in degrees:

    poly = PolynomialFeatures(degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    y_pred = model.predict(X_test_poly)
    r2 = r2_score(y_test, y_pred)

    print(f"Degree {degree}: R-squared = {r2:.4f}")
```
